Outil pour parser un texte selon une chaine de caractère

```python
def get_position(uni_marqueur,chaine):
    out = 0
    if uni_marqueur in chaine : 
        out = chaine.find(uni_marqueur)
    return out
        
def parse_it(marqueur,chaine):
    L = []
    uni_marqueur = "£"
    chaine = chaine.replace(marqueur,uni_marqueur)
    while uni_marqueur in chaine :
        p = get_position(uni_marqueur, chaine)
        L.append(chaine[:p])
        chaine = chaine[p+1:]
    L.append(chaine)
    return L 
```

Module pour trouver la référence TXT de la règle

```python
def represents_int(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
```
